```python
import torch
import transformers
from transformers import AutoModel, AutoTokenizer, pipeline, AutoModelForSequenceClassification
```

```python
model_list  = ['MoritzLaurer/mDeBERTa-v3-base-mnli-xnli', 'facebook/bart-large-mnli', 'joeddav/xlm-roberta-large-xnli', 'sileod/deberta-v3-base-tasksource-nli']
```

```python
file_path = './article.txt'

with open(file_path, 'r') as f:
    loaded_article = f.read()
```

```python
candidate_labels = ['Automotive News', 'Weather News']
```

```python
result_dict = {}

for model in model_list:
    classifier = pipeline("zero-shot-classification",
                      model=model, use_fast=False)

    output = classifier(loaded_article, candidate_labels, max_length=800)
    
    result_dict[model] = output
```

```python
checkpoint = model_list[0]
```

```python
tokenizer = AutoTokenizer.from_pretrained(checkpoint, use_fast=False)
```

```python
inputs = tokenizer(loaded_article, padding=True, truncation=True, return_tensors="pt")
```

```python
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
```
